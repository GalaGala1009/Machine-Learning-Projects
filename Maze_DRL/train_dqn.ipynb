```python
import numpy as np
import scipy.special as sp

from IPython.display import display, clear_output
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.optim as optim
from env.env import MazeEnv
import utils.maze_generator as maze_generator
```

```python
maze = maze_generator.generate_maze(11, 11)

initial_position = [1,1]
goal = [len(maze)-2, len(maze)-2]

maze_env = MazeEnv(maze, initial_position, goal)
```

```python
maze_env.draw()
```

```python
from agents.dqn_agent import DQN_Agent
dqn_agent = DQN_Agent(maze = maze_env,
              buffer_capacity = 500,
              buffer_start_size = 300,
              use_softmax = True
             )

loss_log = dqn_agent.train(1000)
```

```python
plt.plot(loss_log, label = 'Loss', color = 'cornflowerblue')
plt.xlabel('Epoch')
plt.ylabel('MSE')
plt.legend()
plt.show()
```

```python
dqn_agent.test()
```
